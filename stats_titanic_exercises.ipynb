```python
total_surv = df.Survived.sum()
total_surv
fares = []
surv = []
df_sorted = df.sort_values('Fare')
for qnt in range(20):
    qnt = qnt/20 + 0.05
    last_fare_qnt= df_sorted.Fare.quantile(q=qnt - 0.05)  
    fare_qnt = df_sorted.Fare.quantile(q=qnt)  

    last_total = len(df_sorted[df_sorted['Fare'] <= last_fare_qnt])
    new_total =  len(df_sorted[df_sorted['Fare'] <= fare_qnt])
    total_in_qnt = new_total - last_total
   
    last_survived =  df_sorted[df_sorted['Fare'] <= last_fare_qnt].Survived.sum()
    new_survived = df_sorted[df_sorted['Fare'] <= fare_qnt].Survived.sum()
    survived_in_qnt = new_survived - last_survived
    
    surv_in_qnt_pct = survived_in_qnt / total_in_qnt
  
    
    print ("{0:.2f}".format(qnt), fare_qnt, last_survived, new_survived, survived_in_qnt, last_total, new_total, 
          total_in_qnt, surv_in_qnt_pct)
    
    #print("{0:.2f}".format(qnt), fare_qnt, surv_in_qnt_pct)
    fares.append(fare_qnt)
    surv.append(surv_in_qnt_pct)

#plot_data    
plt.figure(figsize=(17,5))
plt.plot(fares, surv)
plt.xlabel("fare")
plt.ylabel("survival rate")

plt.show()
```

```python
# video solutioin that shows the coding style of the solution provider a technique he calls creating
# predicate, which is a sort of subset of the main dataframe

# function takes a criteria that subsets or filters the df, then returns the number of survivors over total
def survival_ratio(predicate):
    series = df[predicate]
    return len(series[series['Survived'] == True]) / len(series)

# creates the series we will populate by simply giving a name, similar to what creating empty list
below_at_median = pd.Series(name='Surv. below/at the median')
above_median = pd.Series(name='Surv. above the median')

# we now pop the series by using a label and value (similar to a Dict)
below_at_median['Age'] = survival_ratio(df['Age'] <= df['Age'].median())
below_at_median['Ticket Price'] = survival_ratio(df['Fare'] <= df['Fare'].median())
below_at_median['Siblings/Spouses'] = survival_ratio(df['SibSp'] <= df['SibSp'].median())
below_at_median['Parents/Children'] = survival_ratio(df['Parch'] <= df['Parch'].median())

above_median['Age'] = survival_ratio(df['Age'] > df['Age'].median())
above_median['Ticket Price'] = survival_ratio(df['Fare'] > df['Fare'].median())
above_median['Siblings/Spouses'] = survival_ratio(df['SibSp'] > df['SibSp'].median())
above_median['Parents/Children'] = survival_ratio(df['Parch'] > df['Parch'].median())

# take the series and create our dataframe using series as rows and then transposing, add col of diff and display
survival_median = pd.DataFrame([below_at_median, above_median],
                               columns=['Age', 'Ticket Price', 'Siblings/Spouses', 'Parents/Children']).transpose()
survival_median['Surv. above - Surv. at/below'] = above_median - below_at_median
survival_median
```

**Find the probability that:**
- A passenger survived
- A passenger was male
- A passenger was female and had at least one sibling or spouse on board
- A survivor was from Cherbourg
- A passenger was less than 10 years old
- A passenger was between 25 and 40 years old
- A passenger was either younger than 20 years old or older than 50

```python
df.head()
df[(df['Sex'] == 'female') & (df['SibSp'] > 0)]
```

```python
def calc_prob(predicate):
    return len(df[predicate]) / len(df)
    

probs = pd.Series()
probs['Survived'] = calc_prob(df['Survived'] == 1)
probs['Male'] = calc_prob(df['Sex'] == 'male')
probs['Female with Sibling or Spouse'] = calc_prob((df['Sex'] == 'female') & (df['SibSp'] != 0))
probs['Survived and from Cherbourg'] = calc_prob((df['Survived'] == 1) & (df['Embarked'] == 'C'))
probs['Age less than 10'] = calc_prob(df['Age'] < 10)
probs['Age between 25 and 40'] = calc_prob((df['Age'] > 25) & (df['Age'] < 40))
probs['Age less than 20 and greater than 50'] = calc_prob((df['Age'] < 20) | (df['Age'] > 50))


probs
```

**Knowing nothing else about the passengers aside from the survival rate of the population (see question above), if I choose 100 passengers at random from the passenger list, whatâ€™s the probability that exactly 42 passengers survive?**

```python
stats.binom.pmf(k=42, n=100, p=probs['Survived'])
```

**Whatâ€™s the probability that at least 42 of those 100 passengers survive?**

```python
fortytwo_plus = 1 - stats.binom.cdf(k=41, n=100, p=probs['Survived'])
fortytwo_plus
```

**Take random samples of 100 passengers and find out how many you need before the fraction of those samples where at least 42 passengers survive matches the probability you calculated previously (within Î”pâ‰ˆ0.05).**

Answers will vary based on chosen seeds. What would happen if you drew every sample with the same seed?

Plot the survival fraction vs the number of random samples.

```python
surv_frac = []
samples = 0
fortytwo_or_more = 0
tol = 1
seed = 7
while tol > 0.05:
    samples += 1
    samp = df.Survived.sample(n=100, random_state=seed).sum()
    if samp >= 42:
        fortytwo_or_more += 1
    tol = abs(fortytwo_plus - fortytwo_or_more / samples)
    print(tol, samples, fortytwo_or_more / samples)    
    seed = seed + 7
    surv_frac.append(fortytwo_or_more / samples)
surv_frac = pd.Series(surv_frac)
ax = surv_frac.plot(y='cum surv frac', figsize=(16,5))
ax.set_xlabel("Number of samples")
ax.set_ylabel('survival fraction')
```

**Did survivors pay more for their tickets than those that did not? Use a 95% confidence level.**

```python
fare_survivors = df.Fare[df['Survived'] == 1]
fare_nonsurvivors = df.Fare[df['Survived'] == 0]
fare_survivors.mean(), fare_nonsurvivors.mean()
two_sided = stats.ttest_ind(fare_survivors, fare_nonsurvivors, equal_var=False, nan_policy='omit')
# one sided test means we need to divide p value by 2 - still well below 0.05
print('t = ', two_sided.statistic)
print('1 sided p = ', two_sided.pvalue/2)
```
